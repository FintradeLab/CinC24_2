# **0. Libraries**

```python
import numpy as np          # Algebraic operations
import pandas as pd         # Manage the data base
import plotly.express as px # Graphics
```

# **3. Cleaning**

## 3.1. Column names

```python
# Time to year
d.rename(columns={'Time': 'year'}, inplace=True)

# Country name to country
d.rename(columns={'Country Name': 'country'}, inplace=True)

# Country code to country_code
d.rename(columns={'Country Code': 'code'}, inplace=True)

# Population, total to pop
d.rename(columns={'Population, total [SP.POP.TOTL]': 'pop'}, inplace=True)

# Poverty headcount to pov
d.rename(columns={'Poverty headcount ratio at $2.15 a day (2017 PPP) (% of population) [SI.POV.DDAY]': 'pov'}, inplace=True)

# Educational attainment, at least Bachelor
d.rename(columns={'Educational attainment, at least Bachelor\'s or equivalent, population 25+, total (%) (cumulative) [SE.TER.CUAT.BA.ZS]': 'edu'}, inplace=True)

# GDP per capita to gdp_pc
d.rename(columns={'GDP per capita, PPP (constant 2021 international $) [NY.GDP.PCAP.PP.KD]': 'gdp_pc'}, inplace=True)
```

```python
d.info()
```

```python
# Duplicate the dataset
d2 = d
```

```python
# Create a new column with year as str
d2['year_str'] = d2['year'].astype(str)
d2.info()
```

## 3.2. Delete columns

```python
d2 = d2.drop(columns=['Time Code'])
d2.info()
```

## 3.3. Delete NaN

```python
# Replace by NaN
d2 = d2.replace('..', np.nan)
d2.head()
```

```python
d2.info()
```

```python
# Drop NaN
d2 = d2.dropna()
d2.info()
```

# **4. Indexing**

Categories of the dataset

```python
# Indexing
d2 = d2.set_index(['country', 'year'])
d2.head()
```
