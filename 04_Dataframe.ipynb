# **3. Cleaning**

## 3.1. Column names

```python
# Time to year
d.rename(columns={'Time': 'year'}, inplace=True)

# Country name to country
d.rename(columns={'Country Name': 'country'}, inplace=True)

# Country code to country_code
d.rename(columns={'Country Code': 'code'}, inplace=True)

# Population, total to pop
d.rename(columns={'Population, total [SP.POP.TOTL]': 'pop'}, inplace=True)

# Poverty headcount to pov
d.rename(columns={'Poverty headcount ratio at $2.15 a day (2017 PPP) (% of population) [SI.POV.DDAY]': 'pov'}, inplace=True)

# Educational attainment, at least Bachelor
d.rename(columns={'Educational attainment, at least Bachelor\'s or equivalent, population 25+, total (%) (cumulative) [SE.TER.CUAT.BA.ZS]': 'edu'}, inplace=True)

# GDP per capita to gdp_pc
d.rename(columns={'GDP per capita, PPP (constant 2021 international $) [NY.GDP.PCAP.PP.KD]': 'gdp_pc'}, inplace=True)
```

```python
d.info()
```

```python
# Duplicate the dataset
d2 = d
```

```python
# Create a new column with year as str
d2['year_str'] = d2['year'].astype(str)
d2.info()
```

## 3.2. Delete columns

```python
d2 = d2.drop(columns=['Time Code'])
d2.info()
```

## 3.3. Delete NaN

```python
# Replace by NaN
d2 = d2.replace('..', np.nan)
d2.head()
```

```python
d2.info()
```

```python
# Drop NaN
d2 = d2.dropna()
d2.info()
```

# **6. Filters and subsets**

## 6.2. Pivots

```python
pivot_table = pd.pivot_table(
    d2,
    values=['pov', 'edu'],  # Analyze both poverty and education
    index='country',  # Group by country
    columns='year_str',  # Separate by year as string
    aggfunc='mean'  # Calculate the average
)
pivot_table.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
# Create the heatmap
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(pivot_table.loc[:, 'pov'], annot=True, cmap='viridis', fmt=".2f")
plt.title('Poverty Levels by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()

# You can create a similar heatmap for 'edu' as well:
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(pivot_table.loc[:, 'edu'], annot=True, cmap='viridis', fmt=".2f")
plt.title('Education Levels by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

```python
# Create the pivot table for population
pop_pivot = pd.pivot_table(d2, values='pop', index='country', columns='year_str', aggfunc='mean')

# Define custom colormap
colors = [(1, 0, 0), (0, 1, 0)]  # Red to Green
cmap = mcolors.LinearSegmentedColormap.from_list('custom_cmap', colors)

# Create the heatmap
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(pop_pivot, annot=False, cmap=cmap, fmt=".0f")  # Remove annotations (annot=False)
plt.title('Population by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```
